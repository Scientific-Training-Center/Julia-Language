## Differential 


https://docs.sciml.ai/DiffEqDocs/stable/


Additional: https://neuralpde.sciml.ai/stable/

```julia
using DifferentialEquations
f(u, p, t) = 1.01 * u
u0 = 1 / 2
tspan = (0.0, 1.0)
prob = ODEProblem(f, u0, tspan)
sol = solve(prob, Tsit5(), reltol = 1e-8, abstol = 1e-8)

using Plots
plot(sol, linewidth = 5, title = "Solution to the linear ODE with a thick line",
     xaxis = "Time (t)", yaxis = "u(t) (in μm)", label = "My Thick Line!") # legend=false
plot!(sol.t, t -> 0.5 * exp(1.01t), lw = 3, ls = :dash, label = "True Solution!")
```

```julia
using DifferentialEquations, Plots

function lorenz!(du, u, p, t)
    du[1] = 10.0 * (u[2] - u[1])
    du[2] = u[1] * (28.0 - u[3]) - u[2]
    du[3] = u[1] * u[2] - (8 / 3) * u[3]
end


u0 = [1.0; 0.0; 0.0]
tspan = (0.0, 100.0)
p = [10.0, 28.0, 8 / 3]
#prob = ODEProblem(lorenz!, u0, tspan, p)
prob = ODEProblem(lorenz!, u0, tspan)
sol = solve(prob)

plot(sol, idxs = (1, 2, 3))
```

## Integrals
https://juliamath.github.io/QuadGK.jl/stable/

```julia
using QuadGK
function f(x)
    return x^2
end

result, error = quadgk(f, 0, 1)

println("Result: ", result)
println("Error: ", error)
```

```julia
quadgk(x -> exp(-x^2), -Inf, Inf)
```

```julia
using QuadGK

# Define the function
f(x) = 1 / (1 - x)

# Set the limits
a = 0
b = 2

# Compute the integral
result, err = quadgk(f, a, b)

# Print the result
println("The integral of 1/(1-x) from $a to $b is: $result")
```
